```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = "retina"
```

```python
def reduce(number):
    str_number = str(number)
    if len(str_number) != 1:
        return reduce(sum([int(v) for v in str_number]))
    else:
        return number
```

```python
x = 12
ans = (((x + 7) * 3) ** 2 - 3) * 2 - 8
```

```python
letters = "abcdefghijklmnop"
```

```python
reductions = [reduce(x) for x in range(50)]
plt.plot(reductions, marker="o")
```

```python
G1 = nx.Graph()

range_numbers = range(150)
for number in range_numbers:
    str_number = str(number)
    if len(str_number) == 1:
        G1.add_edge(number, 0)
    else:
        while len(str_number) != 1:
            reduce_num = sum([int(v) for v in str_number])
            G1.add_edge(number, reduce_num)
            number = reduce_num
            str_number = str(number)
```

```python
values = [reduce(number) for number in range_numbers]
fig, ax = plt.subplots(figsize=(13, 10))
np.random.seed(31415)
nx.draw(G1, with_labels=True, node_color=values, cmap="tab20")
```

## Generando la ecuacin del mago

La ecuacin original del mago est dada por 

$$
    868 + 252 x + 18 x^2
$$

```python
xrange = np.arange(1, 100, 5)
def magic_f(x):
    return 18 * x ** 2 + 252 * x + 868
```

```python
magic_f(2)
```

```python
plt.plot(xrange, magic_f(xrange), marker="o")
```
